```python
import urllib
import bs4 as bs
import re
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/India').read()
```

```python
source
```

```python
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
soup
```

```python
text =""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
text = re.sub(r'\[[0-9]*\]', ' ', text)
```

```python
text
```

```python
text = re.sub(r'\s+', ' ', text)
```

```python
text = text.lower()
```
